```python
def get_tokens(text):
    lowers = text.lower()
    #remove the punctuation using the character deletion step of translate
    identify = string.maketrans("","") 
    delEStr = string.punctuation +string.digits 
    cleanLine = lowers.translate(identify,delEStr) #去掉ASCII 标点符号和空格  
    #cleanLine =line.translate(identify,delEStr) #
    #new_text = text.translate(table,string.punctuation)   
    #new_text = new_text.translate(new_text,string.digits)
    #remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)
    #no_punctuation = lowers.translate(remove_punctuation_map)
    tokens = nltk.word_tokenize(cleanLine)
    return tokens
```

```python
word_tokens=get_tokens(f)
```

```python
frequency = {}

for filtered_sentence in filtered_sentences:
    match_pattern = re.findall(r'\b[a-z]{3,15}\b', filtered_sentence)
 
    for word in match_pattern:
        count = frequency.get(word,0)
        frequency[word] = count + 1
     
frequency_list = frequency.keys()
words_frequency=[]
for words in frequency_list:
    words_frequency.append(frequency[words])
    print words, frequency[words]
    #words_frequency=frequency[words]
    

```
